###  Task 1 : First Create a New Sub Sheet & Fetch all the Data from the Sheet called "Data Set" where Data from the PE Ratio is not Blank & PE Ratio < 100 using Query Function  and sort the data by Descending of Market Cap Amount.

```python
#=QUERY('Data Set'!A:K, "SELECT * WHERE D IS NOT NULL AND D < 100 ORDER BY B DESC", 1)


# Convert 'PE Ratio' column to numeric
df['PE Ratio'] = pd.to_numeric(df['PE Ratio'], errors='coerce')
# Filter the DataFrame to select rows where PE Ratio is not blank and less than 100
filtered_df = df[df['PE Ratio'].notnull() & (df['PE Ratio'] < 100)]
sorted_df = filtered_df.sort_values(by='Market cap (in Rs Cr.)', ascending=False)
sorted_df.to_excel('task1.xlsx', index=False)
sorted_df.sample(10)
```

### Task 2 :  Now Create 2 Columns of Industry & NSE/BSE Code & Fetch Data from the Second Sheet with Vlookup, Make sure the Vlookup Formula is used with "ARRAYFORMULA" meaning formula will be written only once but will be applicable throughout the column, Also make sure the Formula of Vlookup will be applied only until when data is there in the column, handle this with IF CONDITIONS in that same formula.

```python
# =IF(A1="", "", ARRAYFORMULA(VLOOKUP(A2:A, 'Company Info'!A:B, 2, FALSE)))

industry_df = pd.read_csv("NSEBSE Code.csv", skiprows=1)
industry_df.columns = industry_df.iloc[0]
industry_df = industry_df.drop(0)
industry_df = industry_df.reset_index(drop=True)
# Perform a vlook operation or merge operation with the first DataFrame
merged_df1 = pd.merge(sorted_df, industry_df, on='Stock', how='left')
# Fill missing values in the 'Industry' and 'NSE/BSE Code' columns with empty strings
merged_df1['Industry'] = merged_df1['Industry'].fillna('')
merged_df1['NSE/BSE Code'] = merged_df1['NSE/BSE Code'].fillna('')
# Save the merged DataFrame to a new CSV file
merged_df1.to_excel('task2.xlsx', index=False)
merged_df1.head()
```

```python
merged_df1.shape
```

### Task 3 : Now Create 2 More Columns at the End with the Heading "Bear Mode 1" & "Bear Mode 2" Bear Mode 1 Condition : Use IF Condition to print "Yes" if % Change from 52 Week High > 30 else "No" Bear Mode Condition 2 : Use IF Condition to print "Yes" if % Change from 52 Week Low < 30 else "No"

```python
# Convert 'Current price' and '52 Week Low' columns to numeric
merged_df1[['Current price', '52 Week High', '52 Week Low']] = merged_df1[['Current price', '52 Week High','52 Week Low']].apply(pd.to_numeric, errors='coerce')
merged_df1['% Change from 52 Week High'] = ((merged_df1['Current price'] - merged_df1['52 Week High']) / merged_df1['52 Week High']) * 100
merged_df1['% Change from 52 Week Low'] = ((merged_df1['Current price'] - merged_df1['52 Week Low']) / merged_df1['52 Week Low']) * 100
merged_df1.sample(10)
```

```python
# Bear Mode 1
merged_df1['Bear Mode 1'] = merged_df1['% Change from 52 Week High'].apply(lambda x: 'Yes' if x > 30 else 'No')
# Bear Mode 2
merged_df1['Bear Mode 2'] = merged_df1['% Change from 52 Week Low'].apply(lambda x: 'Yes' if x < 30 else 'No')
# Display the DataFrame to verify the changes
merged_df1.head(5)
```

```python
# Create Stock Status column
merged_df1['Stock Status'] = ''
# Apply conditions
merged_df1.loc[(df['PE Ratio'] < 65) & (merged_df1['Bear Mode 1'] == 'No') & (merged_df1['Bear Mode 2'] == 'No'), 'Stock Status'] = 'Good Stock'
# Display the DataFrame to verify the changes
merged_df1.head(5)
```

### Task 5 : Create a Pivot Table Report as Follows :

- Industry wise Total Count of Companies
- Industry wise Average Market Cap
- Industry wise sum of Market Cap

```python
pivot_table_count = pd.pivot_table(selected_data, 
                                    index='Industry',
                                    values='Stock',
                                    aggfunc='count',
                                    fill_value=0)

# Rename the column for better readability
pivot_table_count.columns = ['Total Count of Companies according to Industry']
# print(pivot_table_count)
pivot_table_count.head(100)
```

```python
pivot_table_count.to_excel('task5.1.xlsx')
```

```python
pivot_table_count2 = pd.pivot_table(selected_data, 
                                    index='Industry',
                                    values='Market cap (in Rs Cr.)',
                                    aggfunc='mean',
                                    fill_value=0)

# Rename the column for better readability
pivot_table_count2.columns = ['Average Market Cap based on Industry']
print(pivot_table_count2)
pivot_table_count2.head()
```

```python
pivot_table_count2.to_excel('task5.2.xlsx')
```

```python
pivot_table_count3 = pd.pivot_table(selected_data, 
                                    index='Industry',
                                    values='Market cap (in Rs Cr.)',
                                    aggfunc='sum',
                                    fill_value=0)

# Rename the column for better readability
pivot_table_count3.columns = ['SUM of Market Cap based on Industry']
print(pivot_table_count3)
pivot_table_count3.head()
```

```python
pivot_table_count3.to_excel('task5.3.xlsx')
```
